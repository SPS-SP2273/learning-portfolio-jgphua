## Exercise 1 (Fill in the blanks)

```python
from matplotlib import pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=0.25, yerr=[0.0, 0.5, 1.0, 1.5, 2.0, 2.5], #edited according to errors
             color="red", linestyle="dashed",
             marker='o', markerfacecolor="yellow", #marker marks the data points
             ecolor="black",capsize=5,
             label="Y values with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right') #for legend to appear at bottom right of plot
plt.show()
```

## Exercise 2 (A simple mathematical plot)  

```python
import numpy as np
```

```python
x = np.linspace(0, 50, 500)
np_x = np.array(x)            
sin_x = np.sin(np_x)
e = np.e
y = e**(-np_x/10) * sin_x

plt.plot(x, y, color='blue', linestyle='solid')
plt.xlabel('x', color="darkorange", fontsize = 15)
plt.ylabel('y', color="darkorange", fontsize = 15)
plt.title("y=e^\frac{-x}{10}sin(x)")
plt.grid(alpha=.25)
```

```python
x = np.linspace(0, 50, 500)
np_x = np.array(x)            
sin_x = np.sin(np_x)
e = np.e
y = e**(-np_x/10) * sin_x

plt.plot(x, y, color='blue', linestyle='solid')
plt.xlabel('x', color="darkorange", fontsize = 15)
plt.ylabel('y', color="darkorange", fontsize = 15)
plt.title(r'y=e$\frac{-x}{10}sin(x)$')
plt.grid(alpha=.25)
```

```python
x = np.linspace(0, 50, 500) #for x values from 0 to 50, and to make the curve smoother
np_x = np.array(x)  #changing to numpy          
sin_x = np.sin(np_x)
e = np.e
y = e**(-np_x/10) * sin_x #as per equation

plt.plot(x, y, color='#2178b4', linestyle='solid') #used hex value from lecture
plt.xlabel('x', color="darkorange", fontsize = 15) #for color to be in darkorange and fontsize 15
plt.ylabel('y', color="darkorange", fontsize = 15)
plt.title(r'$y=e^{-x/10}sin(x)$', fontsize = 20) #mathematical way to display on matplotlib
plt.grid(alpha=.25) #grid with opacity of 25%
plt.ylim(-1,1) #added limits to y axis
plt.show()
```
