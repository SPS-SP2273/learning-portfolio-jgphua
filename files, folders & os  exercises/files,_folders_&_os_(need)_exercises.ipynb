### Task 2

```python
def rename_my_file(old_file_name):
    month = old_file_name.split('-')[0] #copied over same code but changed to old_file_name instead of data
    date_month = old_file_name.split('_')[0]
    date = date_month.split('-')[1]
    city_extension = old_file_name.split('_')[1]
    new_file_name = f"{month_dictionary[month]}-{date}-{city_extension}"
    return new_file_name
```

```python
rename_my_file("apr-12_singapore.txt")
```
