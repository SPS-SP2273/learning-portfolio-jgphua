# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6] #list out points on the x axis
y_1 = [0, 2, 4, 6, 8, 10, 12] #lists out points on the y axis
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values', #creates plot, legend, color and linestyle
         color='red', linestyle='dashed') 
plt.xlabel('x-values') #creates labels for x and y axis
plt.ylabel('y-values')
plt.title('X vs Y') #creates title of plot
plt.grid(alpha= 0.25) #adjust opacity of grid, cannot go beyond 1
plt.legend(loc='upper left') #changes location of legend
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6] #list out points on the x axis
y_3 = [0, 2, 4, 6, 8, 10, 120] #lists out points on the y axis
y_4 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_3, label='Y values', #creates plot, legend, color and linestyle
         color='purple', linestyle='dashdot') 
plt.xlabel('x-values') #creates labels for x and y axis
plt.ylabel('y-values')
plt.title('X vs Y') #creates title of plot
plt.grid(alpha= 1) #adjust opacity of grid, cannot go beyond 1
plt.legend(loc='upper right') #changes location of legend
plt.show() #if add this removes this thing <matplotlib.legend.Legend at 0x19c702adb50>
```

### Things to note

plt.plot(x, y, color='red', linestyle='dashed', linewidth=2) and plt.plot(x, y, c='red', ls='dashed', lw=2) will produce the same graph as color, line style and linewidth can be shortformed

plt.show() should be added at end of code to show the plot especially for non interactive environments (like from command line)

```python
plt.plot(x, y_1, y_2) #plots have default values for styling 
```

```python
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2)
plt.plot(x, y_1, label='Y values',
           color='red', linestyle='dashed') #both forms are acceptable (different lines)
```

order of keywords does not matter(eg color linestyle etc)

## 1.2 Adding another plot

```python
plt.plot(x, y_2,
         label='Y2 values', color='blue')
plt.show()
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue') #add a seperate line in same plot
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75 #errors need to be defined first
#errors can be a list

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error, #addition of error bars
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6]) #array as it is np

x2 = x**2                  # The math stuff
sin_x = np.sin(x) #usage of numpy allows to do more mathematical stuff

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()     
#plot not as cleared due to limitations with sinx so can add another axis 
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6]) #array as it is np

x2 = x**2                  # The math stuff
cos_x = np.cos(x) #usage of numpy allows to do more mathematical stuff

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, cos_x, label='cos(x)',
         color='blue')
plt.legend()                              
```

## We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2, two legends for each axis
```

### Things to note

```python
x = np.linspace(0, 6, 100) #use of this changes it to be more smoother as there was not many data points)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

# 3 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150) #save.fig saves the plot to drive
#can specify format name with extension
#use of dpi to adjust resolution
```

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=10) #changed resolution too tiny, overrides previous save if same format
#file is saved in the folder though can specify path plt.savefig('C://Desktop/simple-plot.png', dpi=150)
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt', #load data from files in folder
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') #adds horizontal line across
```

## 4.2 Styles

```python
plt.style.use('fivethirtyeight') #used existing styles to make graph nicer without manual edit

data = np.loadtxt('Complete_TAVG_daily.txt', #load data from files in folder
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') #adds horizontal line across
```

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
